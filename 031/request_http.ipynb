```python
import requests
```

```python
url = 'http://dados.recife.pe.gov.br/dataset/2676fd74-8b40-4248-a1da-0dc45f176b7b/resource/7c005351-2d14-46cf-b2f0-6622a5553b30/download/ocorrencias2022.csv'

data_requests = requests.get(url)
```

```python
data_requests.status_code
```

```python
data_requests.headers
```

```python
with open('../data/samu_2022', 'wb') as csv_data:
    csv_data.write(data_requests.content)
```
